```python
# Importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading csv
df = pd.read_csv("Car_sales.csv")
```

```python
print(df.shape)
```

```python
df.head(5)
```

```python
for i in range(df.shape[0]):
    #Delete row i if not equal to Dodge, Lexus and Volvo
    if(df["Manufacturer"][i]) != "Dodge" and (df["Manufacturer"][i] != "Lexus") and (df["Manufacturer"][i] != "Volvo"):
        #inplace->It is a boolean which makes the changes in data frame itself if True
        df.drop(i, axis = 0, inplace = True)
```

```python
df
```

```python
plt.figure(figsize = (25, 6))

#barplot1
Sales_in_thousands = df["Sales_in_thousands"]
Model = df["Model"].tolist()
plt.bar(Model, Sales_in_thousands)

plt.xlabel("Car Model")
plt.ylabel("Sales in thousands")
plt.show()

#barplot2
plt.figure(figsize = (25, 6))
Sales_in_thousands = df["__year_resale_value"]
Model = df["Model"].tolist()
plt.bar(Model,Sales_in_thousands)

plt.xlabel("Car Model")
plt.ylabel("year resale value")

plt.show()

#barplot2
plt.figure(figsize = (25, 6))
Sales_in_thousands = df["Price_in_thousands"]
Model = df["Model"].tolist()
plt.bar(Model,Sales_in_thousands)

plt.xlabel("Car Model")
plt.ylabel("Price_in_thousands")

plt.show()
```
